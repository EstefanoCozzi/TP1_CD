```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df1 = pd.read_csv('C:/Users/tiefa/Desktop/TP1 - Ciencia de Datos/str_1.csv', sep=';')
df2 = pd.read_csv('C:/Users/tiefa/Desktop/TP1 - Ciencia de Datos/str_2.csv', sep=';')

#Concatenar los dataframes
df = pd.concat([df1, df2], ignore_index=True)

#Mostrar información del dataframe
print("Informacion general del dataframe:\n")
print(df.info())
print("\nPrimeras filas:\n")
print(df.head())
print("\nDescripción estadistica de los datos:\n")
print(df.describe())

#Contar valores faltantes por columna
print("\nValores faltantes por columna:\n")
print(df.isnull().sum())
```

```python
# Análisis de variables con faltas
print("\nValores únicos por variable con faltas:\n")
for col in ['encryption_used', 'login_attempts']:
    print(f"\nColumna: {col}")
    print(df[col].value_counts(dropna=False))

# Limpieza de valores faltantes

# Imputar login_attempts con la media
df['login_attempts'] = df['login_attempts'].fillna(df['login_attempts'].mean())

# Imputar encryption_used con la categoría 'Faltante'
df['encryption_used'] = df['encryption_used'].fillna('Faltante')

# Confirmar que ya no quedan valores faltantes
print("\nValores faltantes después de la imputación:\n")
print(df.isnull().sum())
```
